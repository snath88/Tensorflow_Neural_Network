```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,50,501)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
df=pd.DataFrame(data=y,index=x, columns=['sign'])
```

```python
df
```

```python
test_precent=0.1
```

```python
test_point=np.round(len(df)-len(df)*test_precent).astype('int32')
test_point
```

```python
train=df.iloc[0:test_point]
train
```

```python
test=df.iloc[test_point:]
```

```python
test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
scale.fit(train)
```

```python
train_scale=scale.transform(train)
```

```python
test_scale=scale.transform(test)
```

```python
test_scale
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
len(train_scale),len(test_scale)
```

```python
length=50  ## at 50 length some meaningfull traing, one cycle of sign curve
batch=1
generator=TimeseriesGenerator(train_scale,train_scale,length=length,batch_size=batch)
```

```python
len(generator)
```

```python
x,y=generator[0]
x,y
```

```python
train_scale.shape,test_scale.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
n_features=1
```

```python
model=Sequential()
```

```python
# define model
model = Sequential()

# Simple RNN layer
model.add(SimpleRNN(50,input_shape=(length, n_features)))

# Final Prediction
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit(generator,epochs=5)
```

```python
df_loss=pd.DataFrame(model.history.history)
```

```python
df_loss
```

```python
df_loss.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',patience=2)
```

```python
length=49
n_features=1
generator=TimeseriesGenerator(train_scale,train_scale,length=length,batch_size=n_features)
validation_generator=TimeseriesGenerator(test_scale,test_scale,length=length,batch_size=n_features)  

## generator length should always be 1 less than data size
```

```python
model=Sequential()
model.add(LSTM(50,input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(generator,epochs=20,validation_data=validation_generator,callbacks=[early_stop],workers=5,use_multiprocessing=True)
```

```python
df_loss=pd.DataFrame(model.history.history)
```
