```python
### removing 1perc outliers from price
df_filter=df[df['price']<=df['price'].quantile(0.99)]
```

```python
sns.scatterplot(df_filter,x='long',y='lat',hue='price',alpha=0.2,palette='RdYlGn')
```

```python
sns.boxplot(df_filter,x='waterfront',y='price')
```

```python
df_filter['date']=pd.to_datetime(df_filter['date'])
```

```python
df_filter['year']=df_filter['date'].dt.year
df_filter['month']=df_filter['date'].dt.month
```

```python
### another method to extract year/month
df_filter['date'].apply(lambda x: x.year)
```

```python
df_filter.head(3)
```

```python
### see yearwise price

df_filter.groupby('year').mean()['price'].plot()
```

```python
df_filter.drop('date',axis=1,inplace=True)
```

```python
df_filter.columns
```

```python
df_filter['zipcode'].value_counts()
###total unique 70 zipcode, so wecan drop this
```

```python
df_filter.drop('zipcode',axis=1,inplace=True)
```

```python
X=df_filter.drop('price',axis=1)
y=df_filter['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model=Sequential()
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))

model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')  #mean square error
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=450,batch_size=128,verbose=1,workers=5,use_multiprocessing=True) 
### validation data validate the test data as well
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```
