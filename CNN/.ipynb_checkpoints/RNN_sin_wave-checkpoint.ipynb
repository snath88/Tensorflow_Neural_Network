```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,50,501)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
df=pd.DataFrame(data=y,index=x, columns=['sign'])
```

```python
df
```

```python
test_precent=0.1
```

```python
test_point=np.round(len(df)-len(df)*test_precent).astype('int32')
test_point
```

```python
train=df.iloc[0:test_point]
train
```

```python
test=df.iloc[test_point:]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
train_scale=scale.fit_transform(train)
```

```python
test_scale=scale.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length=50  ## at 50 length some meaningfull traing, one cycle of sign curve
batch=1
generator=TimeseriesGenerator(data=train_scale,targets=train_scale,length=length,batch_size=batch)
```

```python
len(generator)
```

```python
x,y=generator[0]
x,y
```

```python
train_scale
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
n_features=1
```

```python
model=Sequential()
```

```python
model.add(SimpleRNN(50,input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse') ##mse for continuous value
```

```python
model.summary()
```

```python
model.fit(generator,epochs=5)
```

```python
df_loss=pd.DataFrame(model.history.history)
```

```python
df_loss.plot()
```

```python
first_eval_batch=test_scale[-length:]   #first training bath is the last 50 data of test test
```

```python
first_eval_batch.shape, type(first_eval_batch)
```

```python
first_eval_batch=first_eval_batch.reshape((1,length,n_features))
```

```python
model.predict_on_batch(first_eval_batch)
```
